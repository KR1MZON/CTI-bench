```python
import pandas as pd
from bs4 import BeautifulSoup
import requests

req = requests.get("https://attack.mitre.org/groups/")

# Parse the HTML
soup = BeautifulSoup(req.content, 'html.parser')

# Find the table
table = soup.find('table')

# Use pandas to read the table directly from the HTML
df = pd.read_html(str(table))[0]

# Display the DataFrame
print(df)
```

```python
import pandas as pd





# Initialize a set to store unique groups
unique_groups = set()

# Add the 'Name' column values to the set
for name in df['Name']:
    unique_groups.add(name)

# Split 'Associated Groups' by commas, strip spaces, and add each unique element
for associated_groups in df['Associated Groups'].dropna():  # Skip None values
    groups = associated_groups.split(',')
    for group in groups:
        unique_groups.add(group.strip())  # Strip leading/trailing spaces

# Convert the set to a sorted list (optional)
unique_groups = sorted(unique_groups)

# Create a DataFrame from the unique groups
unique_groups_df = pd.DataFrame(unique_groups, columns=['Unique Groups'])

# Save to CSV
unique_groups_df.to_csv('unique_groups.csv', index=False)

print("CSV file with unique groups has been created.")
```
